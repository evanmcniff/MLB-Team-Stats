# Basic stats for understanding

```python
# mean, min, and max wins by team
df.groupby('team_name').agg({'wins': ['mean', 'min', 'max']})
```

```python
# mean, min, and max wins by year
df.groupby('year').agg({'wins': ['mean', 'min', 'max']})
```

```python
# mean, min, and max stolen bases by team
df.groupby('team_name').agg({'stolen_bases': ['mean', 'min', 'max']})
```

```python
# mean, min, and max ERA by team
df.groupby('team_name').agg({'earned_run_average': ['mean', 'min', 'max']})
```

```python
# mean, min, and max ERA by year
df.groupby('year').agg({'earned_run_average': ['mean', 'min', 'max']})
```
